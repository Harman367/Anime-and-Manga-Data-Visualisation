```python
# import required packages
import requests
from bs4 import BeautifulSoup
```

```python
# define the url
url = "https://myanimelist.net/anime/5114/Fullmetal_Alchemist__Brotherhood/stats"
# request the url
r = requests.get(url)
# soup it
soup = BeautifulSoup(r.content,"html.parser")
soup
```

```python
# get rows of the table
rows = soup.find_all("div", {"class":"spaceit_pad"})
# print all rows
for res in rows:
    print(res.text)
```

```python
# get rows of the table
rows = soup.find("div", string="Watching")
# print all rows
print(rows)
```
